```python
def luas_lingkaran(r):
  phi = 3.14
  return phi * r * r

try:
  r = input("Masukkan jari-jari lingkaran: ")
  if not r.strip():
    raise ValueError("Input tidak boleh kosong")
  if not r.isdigit():
    raise ValueError("Hanya boleh memasukkan angka")
  r = int(r)
  print("Luas lingkaran = ", luas_lingkaran(r))
except ValueError as e:
  print(f"Error: {e}")


def luas_persegi(s):
  return s * s

try:
  s = input("Masukkan sisi persegi: ")
  if not s.strip():
    raise ValueError("Input tidak boleh kosong")
  if not s.isdigit():
    raise ValueError("Hanya boleh memasukkan angka")
  s = int(s)
  print("Luas persegi = ", luas_persegi(s))
except ValueError as e:
  print(f"Error: {e}")


def luas_persegipanjang(p, l):
  return p * l

try:
  p = input("Masukkan panjang persegi panjang: ")
  l = input("Masukkan lebar persegi panjang: ")
  if not p.strip():
    raise ValueError("Input tidak boleh kosong")
  if not l.strip():
    raise ValueError("Input tidak boleh kosong")
  if not p.isdigit():
    raise ValueError("Hanya boleh memasukkan angka")
  if not l.isdigit():
    raise ValueError("Hanya boleh memasukkan angka")
  p = int(p)
  l = int(l)
  print("Luas persegi panjang = ", luas_persegipanjang(p, l))
except ValueError as e:
  print(f"Error: {e}")
```
